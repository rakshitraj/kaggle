# Spotify data

## Line charts

```python
spotify_filepath = "./input/data-visualization/spotify.csv"

#Read the data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
spotify_data.describe().astype("int64")
```

```python
# line chart showing daily global streams of each song
plt.figure(figsize=(15,5))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Lineplot
sns.lineplot(data=spotify_data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
```

```python
for song in spotify_data.columns:
    plt.figure(figsize=(10,5))
    plt.title("Daily global stream of {} in 2017-2018".format(song))
    plt.ylabel("Daily plays (Billion)")
    sns.lineplot(data=spotify_data[song], label=song)
```

# Flight Data

```python
flight_filepath = "./input/data-visualization/flight_delays.csv"

# Read the data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
# print data
flight_data
```

## Bar graph

```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Sprit Airlines Flights, by month")

# Bar chart
sns.barplot(x=flight_data.index, y=flight_data['NK']) # NK is coke for spirit airlines

# Add label for y-axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
for airline in flight_data.columns:
    plt.figure(figsize=(8,6))
    plt.title("Average Arrival Delay for Airline code: {} by month".format(airline))
    plt.ylabel("Arrival Delay (in minutes)")
    sns.barplot(x=flight_data.index, y=flight_data[airline])
```

## Heatmap

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by month")
plt.xlabel("Airline Code")
# Heatmap
sns.heatmap(data=flight_data, annot=True)
```

# Insurance data

```python
# Read the data
insurance_data = pd.read_csv("./input/data-visualization/insurance.csv")
```

```python
insurance_data.head()
```

## Scatter Plot

```python
# Create a scatter plot for insurance charges aganist BMI
plt.figure(figsize=(6,6))
sns.scatterplot(x=insurance_data.bmi,
               y=insurance_data.charges)
```

## Regression line

```python
sns.regplot(x=insurance_data['bmi'],
           y=insurance_data['charges'])
```

## Colour coded sctter plots

```python
sns.scatterplot(x=insurance_data['bmi'],
                y=insurance_data['charges'],
                hue=insurance_data['smoker'])
```

## Multiple regression lines - lmplot

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

## Swarmplot - categorical sctter plot

```python
sns.swarmplot(x=insurance_data['smoker'],
             y=insurance_data['charges'])
```

```python
sns.swarmplot(data=insurance_data, x='sex', y='charges', hue='smoker')
```

## Pairplot

```python
plt.figure(figsize=(20,20))
sns.pairplot(data=insurance_data)
```
