```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
spotify_data = pd.read_csv('./input/data-visualization/spotify.csv',
                          index_col="Date",
                          parse_dates=True)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style("dark")
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style("white")
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style("ticks")
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```
